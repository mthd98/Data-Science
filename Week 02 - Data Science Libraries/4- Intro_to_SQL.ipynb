```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('sample.db')
```

## SELECT & FROM


`SELECT (columns_names) FROM table_name`

```python
cursor = con.execute("SELECT ArtistId, Name FROM artists;")

for row in cursor:
  print(row)
```

```python
cursor = con.execute("SELECT Name, AlbumId, GenreId, Composer FROM tracks;")

for row in cursor:
  print(row)
```

```python
cursor = con.execute("SELECT * FROM tracks;")

for row in cursor:
  print(row)
```

## WHERE (AND, OR, NOT, IN, LIKE, etc.)

```python
df = pd.read_sql_query("""
SELECT * FROM albums
WHERE ArtistId = 1;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * FROM customers;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * FROM customers
WHERE Email LIKE '%a_';
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * from invoices
WHERE Total < 15 AND Total > 10;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * from invoices
WHERE Total BETWEEN 10 AND 15;
""", con)
df
```

## Count, Avg, Sum, Min, Max

```python
df = pd.read_sql_query("""
SELECT * FROM invoices;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * FROM invoices
ORDER BY Total ASC
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * FROM invoices
ORDER BY BillingCity ASC
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT AVG(Total), SUM(Total), COUNT(Total), MIN(Total), MAX from invoices;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT * from invoices;
""", con)
df.Total.describe()
```

## JOINS

```python
df = pd.read_sql_query("""
SELECT * FROM albums
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT artists.ArtistId, artists.Name, albums.Title, albums.AlbumId
FROM artists
INNER JOIN albums ON artists.ArtistId = albums.ArtistId
ORDER BY artists.Name ASC;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT artists.ArtistId, artists.Name, albums.Title, albums.AlbumId
FROM artists
LEFT JOIN albums ON artists.ArtistId = albums.ArtistId
ORDER BY artists.Name ASC;
""", con)
df
```

```python
df = pd.read_sql_query("""
SELECT customers.FirstName, customers.LastName, employees.FirstName, employees.LastName
FROM employees
Left JOIN customers ON customers.SupportRepId = employees.EmployeeId;
""", con)
df
```
