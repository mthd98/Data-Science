# Visualizing statistical relationships
Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

We will discuss the `relplot()` seaborn functions in this tutorial. 

For more information about relplot places check the [link.](https://seaborn.pydata.org/generated/seaborn.relplot.html)

## Relating Variables with Scatter Plots
The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

There are several ways to draw a scatter plot in seaborn. The most basic, which should be used when both variables are numeric, is the `scatterplot()` function. In the categorical visualization tutorial, we will see specialized tools for using scatterplots to visualize categorical data. The `scatterplot()` is the default `kind` in `relplot()` `(it can also be forced by setting kind="scatter")`.


```python
sns.relplot(data=dataset,x='BMI',y='SleepTime')
```

We can splitting the figure into subplots depending on one col using the `col` parameter.

```python
plt.figure(figsize = (40,30))
p=sns.relplot(data=dataset,x='BMI',y='SleepTime', col="Sex",)
```

Seaborn gave you the ability to colouring by a discrete variable using the `hue` parameter 


```python
plt.figure(figsize = (40,30))
p=sns.relplot(data=dataset,x='BMI',y='SleepTime', col="AgeCategory", hue="HeartDisease")
```

```python
# Create the year, month and day columns 
def create_dates_col(data):
    # Create empty list for the day, month and year
    day=[]
    month =[]
    year=[]
    # loop over the date column 
    for date in data.Date:
        # splitting the date using '/'
        date = date.split('/')
        # append the day , month and the year to the empty list
        day.append(date[1])
        month.append(date[0])
        year.append(date[2])
    # create the new columns in the dataframe 
    data['Day'] = day
    data['Month'] = month
    data['Year'] = year
    return data
```

```python
# create the day,month and year columns
dataset = create_dates_col(dataset)
```

```python
dataset
```

```python
# group the dataframe by the year column 
group_by_year = dataset.groupby('Year')
```

```python
# plot the 2017 data 
plt.figure(figsize=(20,10))
sns.lineplot(data=group_by_year.get_group('2017'),x='Month',y='Volume')
```

```python
# plot every year data using the relplot as line plot
plt.figure(figsize=(40,40))
sns.relplot(data=dataset,x='Month',y='Volume',kind='line', col='Year', facet_kws=dict(sharex=False), 
          
)
```

# Heatmap 
A heat map is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varies over space.

```python
# Get the data correlations 
corr = dataset.iloc[:,2:].corr()
corr
```

```python
# plot the heatmap for the correlate column
plt.figure(figsize=(10,8))

sns.heatmap(corr,annot=True)
```
